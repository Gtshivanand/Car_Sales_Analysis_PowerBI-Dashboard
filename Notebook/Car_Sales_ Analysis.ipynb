## About the Dataset:
Dataset : https://www.kaggle.com/datasets/missionjee/car-sales-report
The dataset appears to contain car sales records, including details about customers, dealers, car specifications, and pricing. Here are the key columns:

•	Car_id: Unique identifier for each car.

•	Date: Date of sale.

•	Customer Name: Name of the customer.

•	Gender: Gender of the customer.

•	Annual Income: Customer's annual income.

•	Dealer_Name: Name of the car dealer.


•	Company: Car manufacturer (e.g., Ford, Dodge, Cadillac).

•	Model: Specific car model.

•	Engine: Type of engine (e.g., Overhead Camshaft).

•	Transmission: Type of transmission (Auto/Manual).

•	Color: Color of the car.

•	Price ($): Price of the car in dollars.

•	Dealer_No: Unique dealer identification number.

•	Body Style: Type of car body (e.g., SUV, Hatchback).

•	Phone: Dealer's phone number.

•	Dealer_Region: Geographic location of the dealer.

# 3. Model and Company Insights
* What are the top 5 most popular car models?
* Which company has the highest sales volume?
* What is the average price of cars from each company?

```python
#What are the top 5 most popular car models?

top_5_model = df.Model.value_counts().nlargest(5)
print("Top Model:",top_5_model)
```

```python
#Which company has the highest sales volume?

highest_sales_company = df.Company.value_counts().nlargest(1)
highest_sales_company
```

```python
#What is the average price of cars from each company?

avg_price_car_comp =  df.groupby('Company').agg(average_price_car = ('price','mean')).average_price_car.nlargest(10)
avg_price_car_comp.round()
```
